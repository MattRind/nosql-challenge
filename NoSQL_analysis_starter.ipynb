### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': {'$eq': 20}}

# Use count_documents to display the number of documents in the result

number_of_documents = establishments.count_documents(query)
print(f"Number of establishments with a hygiene score of 20 is {number_of_documents}.")
```

```python
# Display the first document in the results using pprint

pprint(establishments.find_one(query))
```

```python
# Convert the result to a Pandas DataFrame
hygiene_score_20 = list(establishments.find(query))
hygiene_score_20_df = pd.json_normalize(hygiene_score_20)

# Print the number of rows in the result
print(f"The number of rows in the Pandas DataFrame showing establishments with a hygiene score \
of 20 is {len(hygiene_score_20_df)}.")

# Display the first 10 rows of the DataFrame

hygiene_score_20_df.head(10)
```

```python
# show Dataframe again with just a few fields unhiding hygiene score column from output above --
hygiene_score_20_df.loc[:,['BusinessName', 'LocalAuthorityName', 'scores.Hygiene', 'scores.Structural']].head()
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.

query = {'LocalAuthorityName': {'$regex': "London"},
                          'RatingValue': {'$gte': 4}  
              }
# Use count_documents to display the number of documents in the result
number_of_documents = establishments.count_documents(query)
print(f"Number of establishments in London with a RatingValue greater than or equal to 4 is {number_of_documents}.\n")

# Display the first document in the results using pprint
pprint(establishments.find_one(query))
```

```python
# Convert the result to a Pandas DataFrame
london_rated_4_or_higher = list(establishments.find(query))
london_rated_4_or_higher_df = pd.json_normalize(london_rated_4_or_higher)

# Display the number of rows in the DataFrame
print(f"Number of rows in the Pandas DF showing eestablishments in London with a \
RatingValue greater than or equal to 4 is {len(london_rated_4_or_higher_df)}.")

# Display the first 10 rows of the DataFrame
london_rated_4_or_higher_df.head(10)
```

```python
# show Dataframe again with just a few fields unhiding RatingValue column from output above --
london_rated_4_or_higher_df.loc[:,['BusinessName', 'LocalAuthorityName', 'RatingValue', 'scores.Hygiene']].head()
```
